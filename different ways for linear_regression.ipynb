```python
import numpy as np
import pandas as pd
from pylab import *
from scipy import stats
import matplotlib.pyplot as plt
```

# lets take some random house in seq feet and their prices.

```python
house=np.array([1,2,3,4,5,6,7])
price=np.array([10000,15000,20000,30000,50000,60000,80000])
plt.plot(house,price,'o')
plt.show()
```

```python
        # this is the function that we gonna use for the regression
slope,intercept,r_value,p_value,std_err=stats.linregress(house,price)
print(slope)
print(intercept)
```

       lets plot it.

```python
plt.plot(house,price,'o')
plt.plot(house,house*slope+intercept,'r--')
plt.show()
```

```python
      # lets say we want to find out what will be the output if house=15
price15=(15*slope+intercept)
print(price15)
```

# here is another way for linear regression using pylab

```python
house=np.array([1,2,3,4,5,6,7])
price=np.array([10000,15000,20000,30000,50000,60000,80000])

p1=polyfit(house,price,1)
   #HERE WE use 1 for a first degree plynomial (i.e linear regression)
print(p1)
plt.plot(house,polyval(p1,house),'r--')
plt.plot(house,price,'o')
plt.show()
```

```python
     # for the second degree polunomial we use 2 and for third 3 so on...
p2=polyfit(house,price,2)
print(p2)
plt.plot(house,polyval(p2,house), 'g-')
plt.plot(house,price,'o')
plt.plot(house,polyval(p1,house),'r--')
plt.show()
```

in the above graph red dotted line is linear regression and green line is the 2nd degree polynomial regression

# here is another way for the regression if we want to write our own algo

```python
#and we gonna the same example above

house=np.array([1,2,3,4,5,6,7])
price=np.array([10000,15000,20000,30000,50000,60000,80000])

def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

m=sum(np.array(de_mean(house))*np.array(de_mean(price)))/sum(np.array(de_mean(house))**2)
c=mean(price)-m*mean(house)
print(m)
print(c)
def house_price(housefloor):
    price=m*housefloor+c
    return price
```

```python
    #lets say we want to find out the price of house=15
house_price(15)
```

```python
plt.plot(house,price,'o')
plt.plot(house,house_price(house),'r--')
plt.show()
```
